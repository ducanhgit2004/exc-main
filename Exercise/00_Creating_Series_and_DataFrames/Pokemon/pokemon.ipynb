### Step 4. Ops...it seems the DataFrame columns are in alphabetical order. Place  the order of the columns as name, type, hp, evolution, pokedex

```python
# Step 4: Reorder columns in a more logical order
pokemon = pokemon[['Name', 'Type 1', 'Type 2', 'HP', 'Attack', 'Defense',
                   'Sp. Atk', 'Sp. Def', 'Speed', 'Total', 'Generation', 'Legendary']]
pokemon.head()

```

### Step 5. Add another column called place, and insert what you have in mind.

```python
# Step 5: Add a new column called 'place'
places = ['Kanto', 'Johto', 'Hoenn', 'Sinnoh', 'Unova', 'Kalos', 'Alola', 'Galar']
pokemon['place'] = [places[i % len(places)] for i in range(len(pokemon))]
pokemon[['Name', 'place']].head()
```
