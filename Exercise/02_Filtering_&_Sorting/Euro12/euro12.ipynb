### Step 1. Import the necessary libraries

```python
# Step 1: Import the necessary libraries
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/Euro_2012_stats_TEAM.csv).

### Step 3. Assign it to a variable called euro12.

```python
# Đọc lại file với đúng delimiter
euro12 = pd.read_csv('/content/euro.tsv', sep=',')
print(euro12.columns.tolist())  # kiểm tra lại tên cột
euro12.head()

```

### Step 4. Select only the Goal column.

```python
# Step 4: Select only the Goals column
euro12['Goals']

```

### Step 5. How many team participated in the Euro2012?

```python
# Step 5: How many teams participated?
print("Number of teams:", euro12['Team'].nunique())
```

### Step 6. What is the number of columns in the dataset?

```python
# Step 6: Number of columns
print("Number of columns:", euro12.shape[1])
```

### Step 7. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline

```python
# Step 7: Select columns and assign to discipline
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
discipline.head()
```

### Step 8. Sort the teams by Red Cards, then to Yellow Cards

```python
# Step 8: Sort by Red Cards, then Yellow Cards
discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=[False, False], inplace=True)
discipline
```

### Step 9. Calculate the mean Yellow Cards given per Team

```python
# Step 9: Calculate mean Yellow Cards per team
print("Mean Yellow Cards per Team:", discipline['Yellow Cards'].mean())
```

### Step 10. Filter teams that scored more than 6 goals

```python
# Step 10: Teams with more than 6 goals
euro12[euro12['Goals'] > 6]
```

### Step 11. Select the teams that start with G

```python
# Step 11: Teams that start with G
euro12[euro12['Team'].str.startswith('G')]
```

### Step 12. Select the first 7 columns

```python
# Step 12: Select first 7 columns
euro12.iloc[:, :7]
```

### Step 13. Select all columns except the last 3.

```python
# Step 13: Select all columns except last 3
euro12.iloc[:, :-3]
```

### Step 14. Present only the Shooting Accuracy from England, Italy and Russia

```python
# Step 14: Shooting Accuracy from England, Italy, Russia
euro12.loc[euro12['Team'].isin(['England', 'Italy', 'Russia']), ['Team', 'Shooting Accuracy']]
```
